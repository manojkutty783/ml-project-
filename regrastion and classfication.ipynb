```python
x=data.drop(columns='Class')
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split as tt
```

```python
x_t,x_te,y_t,y_te=tt(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier as rc
```

```python
mod=rc()
```

```python
mod.fit(x_t,y_t)
```

```python
ypd=mod.predict(x_te)
```

```python
from sklearn import metrics
```

```python
cmc=metrics.confusion_matrix(y_te,ypd)
accc=metrics.accuracy_score(y_te,ypd)
clr=metrics.classification_report(y_te,ypd)
```

```python
print(accc,clr,cmc)
```

```python
import seaborn as ss
```

```python
ss.heatmap(cmc,annot=True)
```
