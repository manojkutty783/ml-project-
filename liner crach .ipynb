```python
class linearegraction:
    def __init__(self):
     self.w=0
     self.b=0
    def pred(self,x):
     ypred=(x*self.w+self.b)
     return ypred
    def mse(self,ypred,y):
        n=len(y)
        mse=(1/n)*sum((y-ypred)**2)
        return mse
    def fit(self,x,y):
        lr=0.005
        it=2000
        n=len(x)
        cost=[]
        for i in range(it):
            ypred=self.w*x+self.b
            dw=(-2/n)*sum(x*(y-ypred))
            db=(-2/n)*sum(y-ypred)
            self.w=self.w-lr*dw
            self.b=self.b-lr*db
            mse=self.mse(ypred,y)
            cost.append(mse)
            
        return cost
            
```

```python
from sklearn.model_selection import  train_test_split as tt
```

```python
x_train,x_test,y_train,y_test=tt(x,y,test_size=0.2,random_state=42)
```

```python
lr=linearegraction()
```

```python
c=lr.fit(x_train,y_train)
print(c)
```

```python
yp=lr.pred(x_test)
```

```python
print(yp)
mse=lr.mse(yp,y_test)
```
