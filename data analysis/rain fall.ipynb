```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

```python
data=pd.read_csv(r"C:\Users\manoj\Documents\ml lab data set\KERLA DATA SET RAINFALL.csv")
```

```python
data.head()
```

```python
data.rename(columns={'Unnamed: 2':'Aug-24.1', 'Unnamed: 4':'Sep-24.1','Unnamed: 6':'Oct 01, 2024 To Oct 17, 2024.1','Unnamed: 7':'Oct 01, 2024 To Oct 17, 2024.2',
                  'Unnamed: 9':'Jun 01, 2024 To Oct 17, 2024 (Water Year).1','Unnamed: 10':'Jun 01, 2024 To Oct 17, 2024 (Water Year).2'},inplace=True)
```

```python
data.head()
data.info()
```

```python

name=data['District'].iloc[1:]
no=data['Aug-24'].values[1:].astype(float).tolist()
#print(data['Aug-24'].values[1:])
print(no)
ach=data['Aug-24.1'].iloc[1:]
#print(no)
ach=list(map(float,ach))
#no=list(map(float,no))
#print(no)
print(ach)


```

```python
x=len(data.index)-1
x=np.arange(x)
pt.figure(figsize=(15,10))

pt.bar(data['District'].iloc[1:],no,width=0.5,label='aug-24 normal(mm)',color='lightblue')
pt.bar(data['District'].iloc[1:],ach,bottom=no,width=0.5,label='aug-24 achuall(mm)')
pt.ylabel('mm rain fall')
pt.xlabel('kerala districts')
pt.title('rain resved in keral districts on aug 2024')
pt.xticks(rotation=65)
pt.legend()
```

```python
oct_no=data['Sep-24'].iloc[1:].astype(float).tolist()
oct_ach=data['Sep-24.1'].iloc[1:].astype(float).tolist()
```

```python
pt.subplots(figsize=(15,10))
pt.bar(data['District'].iloc[1:],oct_no,width=0.5,label='sep-24 normal rain fall')
pt.bar(data['District'].iloc[1:],oct_ach,bottom=oct_no,width=0.5,label='sep-24 achual rain fall')
pt.xticks(rotation=50)
pt.legend()
```
