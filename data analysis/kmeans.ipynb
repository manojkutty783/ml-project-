```python
from sklearn.preprocessing import minmax_scale as mm , StandardScaler as ss
import warnings as wa
wa.filterwarnings('ignore')
```

```python
s=ss()
```

```python
ds=s.fit_transform(data.iloc[:,2:])
```

```python
ds=pd.DataFrame(ds)
```

```python
wss=[]
for i in range(1,10):
    km=KMeans(n_clusters=i,random_state=2)
    km.fit_predict(ds)
    wss.append(km.inertia_)
```

```python
pt.plot(range(1,10),wss)
pt.show()
```

```python
km=KMeans(n_clusters=3)
ypd=km.fit_predict(data.iloc[:,2:])

ks=km.inertia_
print(ypd)
```
